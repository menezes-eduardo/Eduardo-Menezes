```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import math
```

#### Para determinar a posição da bola foi utilizada as seguintes fórmulas:
 $\vec{a} = \dfrac{đv}{đt}$

#### Integrando de ambos os lados teremos
$ a\times{t} + v_{0} = \vec{v} $

#### Integrando a função da velocidade em função do tempo teremos a função da posição
$  \vec{x} = \dfrac{a\times{t^2}}{2} + v\times{t} + x_{0} $

#### Considerando os eixo X e Y e a aceleração como a aceleração da gravidade teremos as seguinte fórmula para a posição
$  \vec{x} = v_{x}\times{t} + x_{0} $

$  \vec{y} = -\dfrac{g\times{t^2}}{2} + v_{y}\times{t} + y_{0} $

```python
# Velocidade
vx = 10*math.cos((math.pi)/6); # Velocidade no eixo x
vy = 10*math.sin((math.pi)/6); # Velocidade no eixo y

# Aceleração da gravidade
ax = 0; # Aceleração no eixo x
ay = -9.81; # Aceleração no eixo y
```

```python
# Vetor tempo
t = np.linspace(0, 1, 25); # Determina 25 valores de tempo entre 0 e 1 segundo
t
```

```python
# Vetor posição
X = np.zeros(len(t)) # Posição no eixo X
Y = np.zeros(len(t)) # Posição do eixo Y
```

```python
for i in range(len(t)):
  X[i] = X[i] + vx*t[i] + ax*(t[i]**2)/2; # Cálculo da posição do eixo X
  Y[i] = Y[i] + vy*t[i] + ay*(t[i]**2)/2; # Cálculo da posição do eixo Y
```

```python
plt.figure(1)
plt.plot(t,X)
plt.title('Gráfico da posição no eixo X em função do tempo')
plt.xlabel('Tempo [s]')
plt.ylabel('Posição no eixo X [m]')

plt.figure(2)
plt.plot(t,Y)
plt.title('Gráfico da posição no eixo Y em função do tempo')
plt.xlabel('Tempo [s]')
plt.ylabel('Posição no eixo Y [m]')
```

```python
plt.figure()
plt.plot(X,Y, marker = 'o', linestyle = '--')
plt.title('Gráfico da posição da bola durante 1 segundo')
plt.xlabel('Posição no eixo X [m]')
plt.ylabel('Posição no eixo Y [m]')
```
